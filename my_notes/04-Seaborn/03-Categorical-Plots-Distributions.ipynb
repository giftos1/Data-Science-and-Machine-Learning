# Categorical Plots
### Distribution within Categories
- Despite distribution plots being able to be used fo single features, it can also be used to compare distributions across categories
- Example: Instead of the distribution of everyone's salary, we can compare the distributions of salaries per level of education.
- Distributions within Categories involve the following plots:
    - Boxplot (most common)
    - Violinplot
    - Swarmplot
    - Boxenplot (Letter-Value Plot)

#### Boxplots
- The Boxplot displays the distribution of a continuous variable through the use of quartiles
- Quartiles separate out the data into 4 equal number of data points:
    - 25% of data points are in bottom quartile.
    - 50th percentile (Q2) is the median
- Boxplot quickly gives statistical distribution information in a visual format
- We can create a box plot per category!

#### Violin plots
 - The violin plot plays a similar role as the box plot.
 - It displays the probability density across the data using a KDE
 - We can imagine it as a mirrored KDE plot
 - The white dot in a violin plot represents the median and the box, the interquartile range
 - The violin plots can also be created per category

#### Swarm plots
- Shows all the data points in the distribution
- For very large data sets, it wont show all the points, but will display the general distribution of them
- They can also be grouped per category
- The size of the points can also be altered

#### Boxen plot (Letter-value plot)
- It is a relatively new plot developed in 2011 by Heike Hofmannm, Karen Kafadar, and Hadley Wickham
- Mainly designed as an expansion upon the normal box plot
- Check out the linked paper later from notes
- NB: It is currently very uncommon. Only use this plot type if you know your audience is familiar with it
- Adv: Using a system of letter-values we can use multiple quantiles instead of strictly quartiles
- NB: The main purpose of data visualizations is to inform, not confuse or show-off various esoteric plots!

## Creating the Plots

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('StudentsPerformance.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(10,4), dpi=200)
sns.boxplot(data=df, y='math score', x='test preparation course', hue='test preparation course')
```

```python
plt.figure(figsize=(10,4), dpi=200)
sns.boxplot(data=df, y='reading score', x='test preparation course', hue='test preparation course')
```

```python
plt.figure(figsize=(10,4), dpi=200)
sns.boxplot(data=df, y='reading score', x='parental level of education', hue='parental level of education')
```

```python
plt.figure(figsize=(10,4), dpi=200)
sns.boxplot(data=df, x='parental level of education', y='reading score', hue='test preparation course')
plt.legend(bbox_to_anchor=(1.05, 0.5))
```

```python
plt.figure(figsize=(10,5), dpi=200)
sns.boxplot(data=df, x='reading score', y='parental level of education', hue='test preparation course')
plt.legend(bbox_to_anchor=(1.05, 0.5))
```

```python
plt.figure(figsize=(10,8), dpi=200)
sns.violinplot(data=df, x='reading score', y='parental level of education', hue='parental level of education')
```

```python
plt.figure(figsize=(10,8), dpi=200)
sns.violinplot(data=df, x='reading score', y='parental level of education', hue='test preparation course')
plt.legend(bbox_to_anchor=(1.05, 0.5))
```

```python
# split combines the plots, 
plt.figure(figsize=(10,8), dpi=200)
sns.violinplot(data=df, x='reading score', y='parental level of education', hue='test preparation course', split=True)
plt.legend(bbox_to_anchor=(1.05, 0.5))
```

```python
# inner='quartile' shows the different quartiles
plt.figure(figsize=(10,8), dpi=200)
sns.violinplot(data=df, x='reading score', y='parental level of education', hue='test preparation course', split=True, inner='quartile')
plt.legend(bbox_to_anchor=(1.05, 0.5))
```

```python
# inner=None removes the white interquatile bar
plt.figure(figsize=(10,8), dpi=200)
sns.violinplot(data=df, x='reading score', y='parental level of education', hue='test preparation course', split=True, inner=None)
plt.legend(bbox_to_anchor=(1.05, 0.5))
```

```python
plt.figure(figsize=(10,8), dpi=200)
sns.violinplot(data=df, x='reading score', y='parental level of education', hue='parental level of education', inner='stick')
# plt.legend(bbox_to_anchor=(1.05, 0.5))
```

```python
# bw stands for bandwidth. The more you increase the lesser the noise (the smoother the plot)
plt.figure(figsize=(10,8), dpi=200)
sns.violinplot(data=df, x='reading score', y='parental level of education', hue='parental level of education', inner='stick', bw_method=0.1)
# plt.legend(bbox_to_anchor=(1.05, 0.5))
```

```python
plt.figure(figsize=(8, 4), dpi=200)
sns.swarmplot(data=df, x='math score', size=2)
```

```python
# Dodge splits out the swarm plots per the secondary hue catgeory
plt.figure(figsize=(8, 4), dpi=200)
sns.swarmplot(data=df, x='math score', y='gender', size=2, hue='test preparation course', dodge=True)
plt.legend(bbox_to_anchor=(1.35, 0.5))
```

```python
sns.boxenplot(data=df, x='math score', y='test preparation course', hue='test preparation course', palette='Dark2')
```

```python
sns.boxenplot(data=df, x='math score', y='test preparation course', hue='gender', palette='Dark2')
plt.legend(bbox_to_anchor=(1.3, 0.5))
```
