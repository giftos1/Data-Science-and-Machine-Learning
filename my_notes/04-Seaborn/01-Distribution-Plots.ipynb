# Distribution Plots
#### Understanding PLOT TYPES
- Distribution plots display a single continuous feature and help visualize properties such as deviation and average values.
- There are 3 main distribution plot types:
    - Rug Plot
    - Histogram
    - KDE Plot
- A rug plot is the simplest distribution plot and merely adds a dash or tick line for every single value.
- The y-axis does not really have a meaning (not interpretable)
- If we count how many ticks there are per various x-ranges, we can create a histogram
- We also choose the number of bins (sections) that we'll use for the rug plots(0-10, 10-20, 20-30)
- Can also create a bar as high as count(number of counts per bins) - histogram
- Y-acis can also be normalized as percent
- Changing number of bins shows more detail instead of general trends


- Seaborn also allows us to add on a KDE plot curve on top of a histogram (Shows what a continuous probabilty function would look like in a particular dataset)
- KDE stands for Kernel Density Estimation.
- It is a method of estimating a probability density function of a random variable
- It is a way of estimating a continuous probability curve for a finite data sample
- KDE plots are best understood by visualizing their 'construction'
- You can change the kernel and bandwith used which can make your KDE show more or less of the variance contained in the data

#### CODING DISTRIBUTION PLOTS WITH SEABORN

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
```

```python
df.head()
```

```python
# height changes the height of the plots 0.5 represents 50%
# plt.figure(figsize=(5, 5), dpi=200)
sns.rugplot(x='salary', data=df, height=0.5) 
```

```python
# DO NOT USE DISTPLOT
sns.displot(data=df, x='salary', bins=20)
```

```python
# Restart and run all while the set command is commented out to reset the styles
# sns.set(style='darkgrid') # edit theme styles
sns.set_theme(style='dark') # dark backrground, NB: it sets the style for every other diagram in the notebook
# sns.set(style='whitegrid') 
sns.displot(data=df, x='salary', bins=10)
```

```python
sns.set(style='ticks') # ticks on y axis
sns.displot(data=df, x='salary', bins=10)
```

```python
sns.displot(data=df, x='salary', bins=10, color='green', edgecolor='lightgray')
```

```python
sns.displot(data=df, x='salary', bins=10, color='green', edgecolor='lightgray', linewidth=4)
```

```python
# Use displot(preferrable), can use histplot too 
sns.set_theme(style='ticks')
sns.histplot(data=df, x='salary')
```

```python
sns.set_theme(style='ticks')
sns.displot(data=df, x='salary', kde=True)
# sns.displot(data=df, x='salary', kde=True, rug=True)
```

```python
sns.kdeplot(data=df, x='salary')
```

```python
np.random.seed(42)
sample_ages = np.random.randint(0, 100, 200)
```

```python
sample_ages
```

```python
sample_ages = pd.DataFrame(sample_ages, columns=['age'])
```

```python
sample_ages
```

```python
sns.rugplot(data=sample_ages, x='age')
```

```python
sns.displot(data=sample_ages, x='age', rug=True, bins=30, kde=True)
```

```python
# bw_adjust determines the amount of ages that are being used in the data. A high value will not show anything meaningful
# fill - fills a colour below the kdeplot
sns.kdeplot(data=sample_ages, x='age', clip=[0,100], bw_adjust=0.2, fill=True) # clips sets the x axis boundaries, [start, stop]
```
