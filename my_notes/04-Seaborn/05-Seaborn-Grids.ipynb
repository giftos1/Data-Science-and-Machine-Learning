# Seaborn grids

- Seaborn grid calls use Matplotlib subplots() to automatically create a grid based off a categorical column.
- Instead of passing in a specific number of cols or rows for the subplots, we can simply pass in the name of the column and seaborn will automatically map the subplots grid.
- Many of seaborn's built-in plot calls are running on top of this grid system.
- Directly calling the grid system allows users to heavily customize plots.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('StudentsPerformance.csv')
```

```python
df.head()
```

```python
sns.catplot(data=df, x='gender', y='math score', kind='box')
```

```python
sns.catplot(data=df, x='gender', y='math score', kind='box', row='lunch')
```

```python
sns.catplot(data=df, x='gender', y='math score', kind='box', hue='lunch')
```

```python
sns.catplot(data=df, x='gender', y='math score', kind='box', col='lunch', row='test preparation course')
```

```python
sns.catplot(data=df, x='gender', y='math score', kind='violin', col='lunch', row='test preparation course')
```

```python
g = sns.PairGrid(data=df, hue='gender')
g = g.map_upper(sns.scatterplot)
g = g.map_diag(sns.histplot)
g = g.map_lower(sns.kdeplot)
g = g.add_legend()
```
