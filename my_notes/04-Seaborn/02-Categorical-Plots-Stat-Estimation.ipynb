# Categorical Plots
#### Statistical Estimation within Categories
- Understanding the Plots
- The Categorical plots below display a statistical metrics per a category.
- E.g. Mean value per category or a count of the number of rows per category.
- It is the visualization equivalent of a groupby() call

- There are two main types of plots for this:
    - countplot()
        - Counts number of rows per category
    - barplot()
        - General form of displaying any chose metric per category.
- The barplot is the general that allows one to choose any measure or estimator for the y axis
- We could plot the mean value and standard deviation per category instead

##### Important Note!
- Be careful with the plots above since the bar is filled and continuous, a viewer may interpret continuity along the y-axis which may be incorrect!
- Always make sure to add additional labelling and explanation for these plots!

## Statistical Estimation within Categories
#### Coding the Plots

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
```

```python
df.head()
```

```python
df['division'].value_counts()
```

```python
plt.figure(figsize=(10,5), dpi=200)
sns.countplot(data=df, x='division', hue='division')
```

```python
plt.figure(figsize=(10,5), dpi=200)
sns.countplot(data=df, x='level of education', hue='level of education')
```

```python
df['level of education'].value_counts()
```

```python
plt.figure(figsize=(10,5), dpi=200)
sns.countplot(data=df, x='level of education', hue='division', palette='Set2')
```

```python
# use the estimator to pass in the numpy function that you want to call
# errorbar='sd' returns a standard deviation
plt.figure(figsize=(10,5), dpi=200)
sns.barplot(data=df, x='level of education', y='salary', hue='level of education', estimator=np.mean, errorbar='sd')
```

```python
plt.figure(figsize=(10,5), dpi=200)
sns.barplot(data=df, x='level of education', y='salary', hue='division', estimator=np.mean, errorbar='sd')
plt.legend(bbox_to_anchor=(1.05, 1)) # position legend outside the graphs
```
