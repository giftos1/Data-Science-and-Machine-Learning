# Comparison plots
#### Understanding the Plots

- They are 2-dimensional plots
- The two main plot types used are:
    - jointplot()
    - pairplot()

#### jointplot()
- We can map histograms to each feature of a scatterplot to clarify the distributions within each feature.
- We can also adjust the scatterplot to be a hex plot or a 2D KDE plot.
- The use of hexagons is useful as the the darker the area the more the points
- 2D KDE plots show shaded distribution between both KDEs

#### pairplot()
- The pairplot() is a quick way to compare all numerical columns in a DataFrame.
- It automatically creates a histogram for each column and a scatterplot comparison between all possible combinations of columns.
- Warning:
    - pairplot() can be CPU and RAM intensive for large DataFrames with many columns
    - It is a good idea to first filter down to only the columns you are interested in.

#### Coding the plots

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('StudentsPerformance.csv')
```

```python
df.head()
```

```python
plt.figure(dpi=200)
sns.jointplot(data=df, x='math score', y='reading score')
```

```python
# kind defines the type of plot to be displayed on the inside
plt.figure(dpi=200)
sns.jointplot(data=df, x='math score', y='reading score', kind='hex')
```

```python
plt.figure(dpi=200)
sns.jointplot(data=df, x='math score', y='reading score', kind='scatter', alpha=0.2)
```

```python
plt.figure(dpi=200)
sns.jointplot(data=df, x='math score', y='reading score', kind='hist')
```

```python
plt.figure(dpi=200)
sns.jointplot(data=df, x='math score', y='reading score', kind='kde', fill=True)
```

```python
plt.figure(dpi=200)
sns.jointplot(data=df, x='math score', y='reading score', hue='gender')
```

```python
# grabs all numerical columns and returns the histograms as well as the relation with each of the combinations
sns.pairplot(data=df)
```

```python
sns.pairplot(data=df, hue='gender')
```

```python
# tougher to read histograms on histograms
sns.pairplot(data=df, hue='gender', diag_kind='hist')
```

```python
# corner gets rid of the duplicates and shows just one corner
sns.pairplot(data=df, hue='gender', corner=True)
```
