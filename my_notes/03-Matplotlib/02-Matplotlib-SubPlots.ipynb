# Matplotlib - Subplots functionality

- Matplotlip comes with a pre-configured function call plt.subplots() that allows to easily create a Figure and Axes object in side by side formations
- The plt.subplots() command returns a tuple containing the Figure canvas and then a numpy array holding the axes objects
- The plt.subplots() command returns a tuple containing the Figure canvas and then a numpy array holding the axes objects
- plt.subplots() returns a tuple which by common convention we label (fig, axes):
    - fig
        - This is the entire Figure canvas.
    - axes
        - This is the numpy array holding each of the axes according to position in the overall canvas


```python
a = np.linspace(0,10,11)
b = a ** 4
```

```python
x = np.arange(0, 10)
y = 2 * x
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(4,10), dpi=200)
axes[0][0].plot(x, y)
axes[0][1].plot(x, y) # bottom right
axes[1][0].plot(x, y)
axes[1,0].set_xlabel('x label 1,0')

axes[1][1].plot(a,b)
axes[1,1].set_title('1,1 TITLE')

fig.suptitle('Figure title', fontsize=20)


# fig.subplots_adjust(wspace=0.9, hspace=0.5) # manual subplots spacing
# axes[1].plot(a, b)

# One dimensional
# fig, axes = plt.subplots(nrows=3, ncols=1)
# axes[0].plot(x,y)

plt.tight_layout() # Ensures no axes are overlapping(automatically)
# for ax in axes:
#     ax.plot(x, y)

fig.set_figwidth(10)
fig.savefig('new_subplots.png', bbox_inches='tight')
```
