# Matplotlib Figure Object
- Understanding The Figure

- The more comprehensive Matplotlib OOP API makes use of a Figure object
- We then add axes to this Figure object and then plot on those axes
- This allows for very robust controls over the entire plot
- When you call plt.figure(), a blank canvas is created. We can then specify dimensions as plt.figure(figsize=(10,10))
- fig = plt.figure()
  - axes = fig.add_axes([0,0,1,1])
  - axes.plot(x,y)
  - 0,0 represent the x,y from the lower left corner of Axes
  - 1,1 represent the width and height of Axes
- In theory we could set axes side by side using plt.figure() calls, but typically it is easier to use plt.subplots() function calls for this

```python
# All these commands should go in the same cell
fig = plt.figure()

# LARGE AXES

axes1 = fig.add_axes([0,0,1,1])
# axes.plot(x, y)
axes1.plot(a, b)
axes1.set_xlim(0, 8)
axes1.set_ylim(0, 8000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title('Power of 4')

# SMALL AXES
axes2 = fig.add_axes([0.2, 0.5, 0.3, 0.3])
axes2.plot(a, b)

axes2.plot(a, b)
axes2.set_xlim(1, 2)
axes2.set_ylim(0, 50)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('Zoomed In')
```
