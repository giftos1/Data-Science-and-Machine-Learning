```python
import numpy as np
import matplotlib.pyplot as plt
```

# Matplotlib Styling
#### LEGENDS

- Matplotlib offers very robust styling functions that allow us to edit colors, legends, line widths, markers, and much more!
- Main stylings used:
    - Legends
    - Visual Styling
        - Colors
        - Editing lines
                - colors, widths, styles
        - Editing Markers
                - Colors, Size, Styles, Edges

```python
x = np.linspace(0, 11, 10)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x, x, label='X vs X')
ax.plot(x, x**2, label='X vs X**2')
# ax.legend(loc='lower left')
# ax.legend(loc='lower right')
ax.legend(loc=(1.1, 0.5)) # manual adjustment
```

# Matplotlib Styling
#### VISUAL STYLING

```python
fig1 = plt.figure()

ax1 = fig1.add_axes([0,0,1,1])

# possible linestyle options '--', '-', '-.', ':', 'steps'

ax1.plot(x, x, color='#32a899', label='X vs X', lw=0.8, marker='.', markersize='5', markerfacecolor='red', markeredgewidth='6', markeredgecolor='orange') # RGB HEX color, lw is thickness, use ls/linestyle for linestyle options

lines = ax1.plot(x, x+1, color='red', label='X vs X+1', linewidth=2.5) # linewidth for thickness
lines[0].set_dashes([5,2,10,2]) # 5 length of a dash, 2 whitespaces, 10 length of a dash, 2 whitespaces
# ax1.legend()
```
